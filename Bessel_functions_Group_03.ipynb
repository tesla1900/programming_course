In this exercise we have to verify two simple functions involving Bessels functions of the first kind.

```python
#import packages 
import scipy.special as ss
import numpy as np
import matplotlib.pyplot as plt
import matplotlib




#for plot parameters 
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.linewidth'] = 2.0


#defining function 
def jn_asym(n,x):
    """Asymptotic form of jn(x) for x>>n"""

    return np.sqrt(2.0 / np.pi / x) * np.cos(x - (n * np.pi / 2.0 + np.pi / 4.0))



#creating a set of values for x
x = np.linspace(0., 50, 500)




# plot J_0, J_1 and J_5.
for n in [0, 1, 5]:
    plt.plot(x, ss.jn(n, x), label='$J_%d$' % (n))

    
#asymptotic behaviour     
x_asym = x[x > n]
plt.plot(x_asym, jn_asym(n, x_asym), linewidth = 2.0,
         label='$J_%d$ (asymptotic)' % n)




#plotting
plt.title('Bessel Functions')
plt.xlabel('x')
plt.ylabel('$J_n(x)$')
plt.legend()
plt.axhline(0)
plt.show()

```

$\textbf{Result: The asymptotic form described above is a very good approximation for the Bessel Function at large values of x}$

```python
#import packages 
import scipy.special as ss
import numpy as np
import matplotlib.pyplot as plt
import matplotlib



#for plot parameters 
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.linewidth'] = 2.0


# recursion relation
#J(n+1,x) = (2n/x)J(n,x)-J(n-1,x), n = 5



#defining a set of values for x
x = np.linspace(0.1, 50, 500)



# construct both sides of the recursion relation, these should be equal
n = 5



# the scipy implementation of jn(5);
j_n = ss.jn(5, x)



# The recursion relation:
j_n_rec = (2.0 * (n - 1) / x) * ss.jn(n - 1, x) - ss.jn(n - 2, x)



#difference between two formulas and plotting
plt.semilogy(x, abs(j_n - j_n_rec), 'r+-', linewidth=2.0)
plt.title('Error in recursion for $J_%s$' % n)
plt.xlabel('x')
plt.ylabel('$|J_n(5) - J_{n,rec}(5)|$')
plt.grid()
plt.show()
```

```python
#import packages 
import scipy.special as ss
import numpy as np
import matplotlib.pyplot as plt
import matplotlib



#for plot parameters 
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['axes.linewidth'] = 2.0




# recursion relation
#J(n+1,x) = (2n/x)J(n,x)-J(n-1,x), n = 5



#defining a set of values for x
x = np.linspace(0.1, 25, 500)



# construct both sides of the recursion relation, these should be equal
n = 10



# the scipy implementation of jn(5);
j_n = ss.jn(10, x)



# The recursion relation:
j_n_rec = (2.0 * (n - 1) / x) * ss.jn(n - 1, x) - ss.jn(n - 2, x)



#difference between two formulas and plotting
plt.semilogy(x, abs(j_n - j_n_rec), 'r+-', linewidth=2.0)
plt.title('Error in recursion for $J_{%s}$' % n)
plt.xlabel('x')
plt.ylabel('$|J_n(10) - J_{n,rec}(10)|$')
plt.grid()
plt.show()
```

$\textbf{Result: We can see that the scipy Bessel Function implementation is satisfied for both the cases (n=5 and n=10) up to the machine number precision of $\textbf{$\approx 10^{-16}$}$ }$
